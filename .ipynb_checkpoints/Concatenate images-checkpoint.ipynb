### This notebook demonstrates how to concatenate images horizontally and vertically.

```python
%matplotlib inline
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = 'inputs/lena.jpeg'
```

```python
img_1 = cv2.imread(img_path)
img_2 = cv2.imread(img_path)
```

```python
img_rgb = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

### Concatenate images vertically.

```python
img_v = cv2.vconcat([img_1, img_2])
cv2.imwrite('inputs/lena_vertical_concat.jpg', img_v)
```

```python
img_rgb = cv2.cvtColor(img_v, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

### Concatenate images horizontally.

```python
img_h = cv2.hconcat([img_1, img_2])
cv2.imwrite('inputs/lena_horizontal_concat.jpg', img_h)
```

```python
img_rgb = cv2.cvtColor(img_h, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
