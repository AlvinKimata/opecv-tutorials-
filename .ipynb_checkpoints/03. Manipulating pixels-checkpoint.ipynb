```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('inputs/flamingo.jpg')
```

```python
#Define a function to display an image with cv2.
def display(name, file):
    plt.imshow(file)
```

```python
display('flamingo', image)
```

```python
#Accessing pixels in the image
img_pixels = image[0:100, 0:100]

display('img pixels', img_pixels)
```

```python
#Manipulating pixels in an image.
green = (0, 255, 0)

image[0:100, 0:100] = green
```

```python
display('manipulated', image)
```
