```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
canvas = np.zeros((300, 300, 3), dtype = 'uint8')
```

```python
#Function to display a canvas.
def display(name, file):
    plt.imshow(file)
    plt.show()
```

```python
display('canvas', canvas)
```

```python
#Green diagonal line.
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
display('canvas', canvas)
```

```python
red = (0, 0, 255)
cv2.line(canvas, (0, 300), (300, 0), red, 2)
display('red line on canvas', canvas)
```

## Drawing rectangles.

```python
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 350), (225, 200), blue, -1)
display('rectangle', canvas)
```

## Circles.

```python
canvas = np.zeros((300, 300, 3), dtype = 'uint8')

#Define x and y centers.
centerx = canvas.shape[1] // 2
centery = canvas.shape[0] // 2

#Color to be displayed.
color = (255, 255, 255)

for radius in range(0, 175, 25):
    cv2.circle(canvas, (centerx, centery), radius, color, 2)
    
display('circle', canvas)
```

```python
cv2.circle(canvas, (centerx, centery), 20, color, -1)
display('circle', canvas)
```
