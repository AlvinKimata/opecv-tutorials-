````python
import cv2
````

## Writing videos.

### The syntax for `VideoWriter` is: 

```python
VideoWriter(filename, apiPreference, fourcc, 
            fps, frameSize[, isColor])
```

VideoWriter_fourcc('M', 'J', 'P', 'G')

````python
cap = cv2.VideoCapture(0)


if (cap.isOpened() == False):
    print('Error reading video from file')

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

size = (frame_width, frame_height)

result = cv2.VideoWriter('../inputs/output_from_cam.avi', 
                         cv2.VideoWriter_fourcc(*'MJPG'),
                         10, size)

while True:
    ret, frame = cap.read()
    if ret == True:
        
        #Write the frames into the avi file.
        result.write(frame)
        
        #Display frame.
        cv2.imshow('frame', frame)
        
        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release()
result.release()
cv2.destroyAllWindows()
````
