```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('inputs/faces.jpg')
```

```python
plt.figure(figsize = (10, 7))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier('inputs/haarcascade_frontalface_default.xml')
```

## Face detection.

```python
#1. Convert image to grayscale.
image = img.copy()

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#2. Apply image to cascade classifier.
bboxes, detections = face_cascade.detectMultiScale2(gray,minNeighbors = 8)
```

```python
bboxes
```

- The first item in the arrays represent the x coordinate of the face detected.
- The second item in the arrays represent the y coordinate of the face detected.
- The third and forth item represent the width and height of the bounding boxes respectively.

```python
detections
```

```python
#3  Drawbounding boxes.
color = (0, 255, 0)
for x, y, w, h in bboxes:
    cv2.rectangle(image, (x, y), (x + w, y + h), color, 1)
    
plt.figure(figsize = (10, 9))
plt.title('Face detection with Haar cascade classifier')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
