```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
canvas = np.zeros((300, 300, 3), dtype = 'uint8')
```

```python
#Define a function to display a canvas.
def display(name, file):
    plt.title(name)
    plt.imshow(file)
```

```python
display('canvas', canvas)
```

```python
#Green diagonal line.
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
display('canvas', canvas)
```

```python
red = (0, 0, 255)
cv2.line(canvas, (0, 300), (300, 0), red, 3)
display('canvas', canvas)
```

## Drawing rectangles

```python
#Draw a rectangle and fill it inside.
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 200), blue, -1)
display('canvas', canvas)
```

## Draw circles.

```python
#Re-initialize the canvas again.
canvas = np.zeros((300, 300, 3), dtype = 'uint8')

#Define the x, y centers.
centerx, centery = (canvas.shape[1] // 2, canvas.shape[0] // 2)
```

```python
#Color to be displayed.
color = (255, 255, 255)

#r is radius.
for r in range(0, 175, 25):
    cv2.circle(canvas, (centerx, centery), r, color, 2)
    
display('circle', canvas)
```

### Abstract drawing.

```python
canvas = np.zeros((300, 300, 3), dtype = 'uint8')

for item in range(0, 25):
    radius = np.random.randint(5, 200) #25 circles with random radii.
    color = np.random.randint(0, 256, size = (3, )).tolist()
    pt = np.random.randint(0, 300, size = (2, )) #2 random points.
    
    cv2.circle(canvas, tuple(pt), radius, color, -1)

display('random circles', canvas)
```

```python
#Save image.
cv2.imwrite('outputs/circle_drawing.png', canvas)
```
