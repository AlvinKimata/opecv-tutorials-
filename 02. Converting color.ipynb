```python
import cv2 
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
image = cv2.imread('inputs/flamingo.jpg')
```

```python
image.shape
```

```python
b, g, r = cv2.split(image)
b.shape
```

```python
cv2.imshow('blue', b)
cv2.imshow('red', r)
cv2.imshow('color', image)
cv2.waitKey(0) #Wait for any key to be pressed in order to close display window(s).
cv2.destroyAllWindows() #Close the window(s)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.imshow('RGB', image_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Grayscale.
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray', image_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Save grayscale image.
cv2.imwrite('inputs/gray flamingo.jpg', image_gray)
```
